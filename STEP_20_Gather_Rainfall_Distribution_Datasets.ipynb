### Define function to download, unzip and load resources into the source file geodatabase

See notes in further cell for dealing with SSL certificate errors.

````python
def fetchResource(data,fgdb):
    
    if 'zipdr' in data:
        zipshp = data['zipdr'] + '/' + data['shp'];
    else:
        zipshp = data['shp'];
        
    if 'query' in data:
        query = data['query'];
    else:
        query = None;
        
    if 'fix' in data:
        fix = data['fix'];
    else:
        fix = data['shp'];
    
    file = os.getcwd() + os.sep + data['file'];

    if not os.path.exists(file):
        
        print("  attempting to download " + str(data['url']));
        r = requests.get(
             url    = data['url']
            ,stream = True
            ,verify = True
        );
        
        with open(file,'wb') as f:
            for chunk in r.iter_content(chunk_size=1024): 
                if chunk:
                    f.write(chunk);
    
    if arcpy.Exists(arcpy.env.scratchFolder + os.sep + data['shp'] + '.shp'):
        arcpy.Delete_management(arcpy.env.scratchFolder + os.sep + data['shp'] + '.shp');
        
    with zipfile.ZipFile(file) as zf:
        
        for ext in ['.dbf','.prj','sbn','.sbx','.shp','.shx']:
            if zipshp + ext in zf.namelist():
                with zf.open(zipshp + ext) as s:
                    with open(arcpy.env.scratchFolder + os.sep + fix + ext,'wb') as t:
                        shutil.copyfileobj(s,t);
        
    if arcpy.Exists(fgdb + os.sep + fix):
        arcpy.Delete_management(fgdb + os.sep + fix);

    print("  processing " + fix);
    arcpy.conversion.FeatureClassToFeatureClass(
         in_features  = arcpy.env.scratchFolder + os.sep + fix + '.shp'
        ,out_path     = fgdb
        ,out_name     = fix
        ,where_clause = query
    );
````

#### SSL Errors

The Python [requests](https://docs.python-requests.org/en/master/) package used in this step obtains missing intermediate certificates from the related [certifi](https://github.com/certifi/python-certifi) package.  This is a static system whereby Python requests can only "fill in" missing certificates with ones provided by certifi.  This contrasts with most web browsers which will silently fetch and cache any missing intermediate certificates using [AIA fetching](https://www.thesslstore.com/blog/aia-fetching/).

As of May 2021, the server(s) underpining www.wcc.nrcs.usda.gov are misconfigured such that they do not provide the correct certificate chain leaving out the key intermediate certificate, "DigiCert TLS RSA SHA256 2020 CA1".  While a web browser will obtain the missing certificate via AIA fetching, Python only has the certifi package to fall back upon.

The good news is this particular certificate was added to the certfi package with the 2020.12.5 release.  If your Python installation has this version of certifi, things will work without problem.  However, the bad news is that the stock Python 3.7.9 provided with ArcGIS Pro 2.7.x only includes certifi version 2020.6.5.

If the following cell returns an error similar to
```
SSLError(SSLCertVerificationError(1, '[SSL: CERTIFICATE_VERIFY_FAILED] certificate verify failed: unable to get local issuer certificate (_ssl.c:1124)')))
```
then its most likely this problem you are encountering.

Some options:

1. Download the individual NRCS file resources manually using your web browser.  Place the downloaded zipfiles into the **files** folder of the project.
 
 
2. In your ArcGIS Pro Python Package Manager, upgrade the certifi package for any of your cloned Python environments.  Then open a Python prompt to that environment (inside or outside Pro) and execute "import certifi".  Its uninituitive but regardless of which clone is active, there is actually only ever one certificate store in existence (located at C:\Users\%USERNAME%\AppData\Local\.certifi\cacert.pem).  There seems to be logic to always retain the most recent certificate store.  Thus even if you delete the cloned environment with the upgraded certifi package, the new certificate store will persist.
 
 
3. Waiting for ArcGIS Pro 2.8 might in fact solve the matter if Esri bundles the newer 2020.12.5 version of certifi.
 
 
4. It may be worth complaining to USDA about their misconfigured server.  However, this type of problem is [fairly widespread](https://docs.google.com/document/d/1ryqFMSHHRDERg1jm3LeVt7VMfxtXXrI8p49gmtniNP0/edit?pli=1) across the internet and as it does not affect web browsers it may not be high on their list of things to fix.
