```python
# Verify target directory resources exists and create if not
resources = os.getcwd() + os.sep + 'resources';
files     = os.getcwd() + os.sep + 'files';

if not arcpy.Exists(resources):
    os.mkdir(resources);
    
```

```python
if arcpy.Exists(resources + os.sep + 'D4EMLite_PREC_Details.txt'):
    arcpy.Delete_management(resources + os.sep + 'D4EMLite_PREC_Details');
  
flds = [
     'StationId'
    ,'DataType'
    ,'FileName'
    ,'Id'
    ,'Lat'
    ,'Long'
    ,'Scenario'
    ,'Constituent'
    ,'SDate'
    ,'EDate'
    ,'YrCount'
    ,'Value'
    ,'StaNam'
];

def tabWriter(fc,outfile,flds,outflds=None):

    if outflds is None:
        outflds = flds;
        
    with open(outfile,'w',newline='',encoding='utf-8') as f:
        writer = csv.writer(f,delimiter='\t',lineterminator='\n');

        with arcpy.da.SearchCursor(fc,flds) as incur:
            writer.writerow(outflds);

            for row in incur:
                if row[0] is not None:
                    writer.writerow(row);

    cnt = arcpy.GetCount_management(outfile)[0];         
    print("  wrote " + str(cnt) + " records to " + os.path.basename(outfile) + ".");

tabWriter(
     fc      = rezfgdb + os.sep + 'D4EMLite_PREC_Details'
    ,outfile = resources + os.sep + 'D4EMLite_PREC_Details.txt'
    ,flds    = flds
);

print("DONE");    
```

```python
if arcpy.Exists(resources + os.sep + 'PREC2035HotDry.txt'):
    arcpy.Delete_management(resources + os.sep + 'PREC2035HotDry.txt');

if arcpy.Exists(resources + os.sep + 'PREC2035Central.txt'):
    arcpy.Delete_management(resources + os.sep + 'PREC2035Central.txt');
    
if arcpy.Exists(resources + os.sep + 'PREC2035WetWarm.txt'):
    arcpy.Delete_management(resources + os.sep + 'PREC2035WetWarm.txt');
    
if arcpy.Exists(resources + os.sep + 'PREC2060HotDry.txt'):
    arcpy.Delete_management(resources + os.sep + 'PREC2060HotDry.txt');

if arcpy.Exists(resources + os.sep + 'PREC2060Central.txt'):
    arcpy.Delete_management(resources + os.sep + 'PREC2060Central.txt');
    
if arcpy.Exists(resources + os.sep + 'PREC2060WetWarm.txt'):
    arcpy.Delete_management(resources + os.sep + 'PREC2060WetWarm.txt');
    
flds = [
     'StationId'
    ,'Jan'
    ,'Feb'
    ,'Mar'
    ,'Apr'
    ,'May'
    ,'Jun'
    ,'Jul'
    ,'Aug'
    ,'Sep'
    ,'Oct'
    ,'Nov'
    ,'Dec'
    ,'Ann'
];

tabWriter(
     fc      = rezfgdb + os.sep + 'PREC2035HotDry'
    ,outfile = resources + os.sep + 'PREC2035HotDry.txt'
    ,flds    = flds
);

tabWriter(
     fc      = rezfgdb + os.sep + 'PREC2035Central'
    ,outfile = resources + os.sep + 'PREC2035Central.txt'
    ,flds    = flds
);

tabWriter(
     fc      = rezfgdb + os.sep + 'PREC2035WetWarm'
    ,outfile = resources + os.sep + 'PREC2035WetWarm.txt'
    ,flds    = flds
);

tabWriter(
     fc      = rezfgdb + os.sep + 'PREC2060HotDry'
    ,outfile = resources + os.sep + 'PREC2060HotDry.txt'
    ,flds    = flds
);

tabWriter(
     fc      = rezfgdb + os.sep + 'PREC2060Central'
    ,outfile = resources + os.sep + 'PREC2060Central.txt'
    ,flds    = flds
);

tabWriter(
     fc      = rezfgdb + os.sep + 'PREC2060WetWarm'
    ,outfile = resources + os.sep + 'PREC2060WetWarm.txt'
    ,flds    = flds
);

print("DONE");
```
